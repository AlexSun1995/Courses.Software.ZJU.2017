```python
import pickle
# 每一项代表一个段落出现的实体
with open('entity_set.pickle','rb') as f:
    entity_word_set = pickle.load(f,encoding='utf-8')
new_set = []
for item in entity_word_set:
    if len(item) > 0:
        new_set.append(list(item))
print(len(new_set))
print(new_set)
```

```python
nodes = set()
for li in new_set:
    for item in li:
        nodes.add(item)
print(len(nodes))
```

```python
# entity_word_set中的每一项list，两两组合成元组，代表边信息
edges = []
for li in new_set:
    size = len(li)
    for i in range(size - 1):
        for j in range(i+1,size):
            edge = (li[i],li[j])
            edges.append(edge)
print(len(edges))
```

```python
print(edges[100:200])
```

```python
import csv
with open('words_edges.csv', 'w',newline='') as f:
    writer = csv.writer(f)
    size = len(edges)
    for i in range(size):
        writer.writerow(list(edges[i]))
    print("edges write success!")
```
