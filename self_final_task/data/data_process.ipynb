```python
#-*-coding:utf-8-*-

import time
# 数据预处理脚本 数据的第一个字段　20111230104333　依次代表了了年，月，时，分，秒　数据，由于数据
# 所提供的时间都是2011年12月30日的数据，此脚本的功能是只提取出时，分，秒数据，并按照\t进行分割
start_time = time.time()
cnt = 0
with open('../sogou.500w.utf8', 'r') as fo:
    with open('../sogou.50w.new','a') as fw:
        for line in fo:
            line.rstrip('\n')
            cnt += 1
            if cnt % 10000 == 0:
                print("%f%% done.." %(cnt*100 / 500000))
            ele_list = line.split('\t')
            all_time = ele_list[0]
            # print(all_time)
            hour = all_time[8:10]
            minute = all_time[10:12]
            sec = all_time[12:14]
            # print("%s %s %s" %(hour,minute,sec))
            fw.writelines([hour,'\t',minute,'\t',sec,'\t'])
            for i in range(1,len(ele_list)):
                fw.write(ele_list[i])
                if i < (len(ele_list) -1):
                    fw.write('\t')
            # fw.write('\n')
end_time = time.time()
print("time spend: %d seconds" %(end_time - start_time))
```
